Ques. Square all numbers using map.

```python
nums = [2, 5, 8, 11, 14, 17, 20, 23, 26, 29]
squared= list(map(lambda x:x**2,nums))
print(squared)
```

Ques. Filter even numbers.

```python
even= list(filter(lambda x:x%2==0,nums))
print(even)
```

Ques. Sort list of tuples by second value.

```python
nums=[("Alice", 85), ("Bob", 78), ("Charlie", 92)]
nums.sort(key=lambda num:num[1])
print(nums)
```

Ques. Convert names to uppercase using map.

```python
names = [
    "Tushar",
    "Rahul",
    "Aman",
    "Priyanka",
    "Rohit",
    "Anjali",
    "Vivek",
    "Karan",
    "Shivani",
    "Aryan"
]
```

```python
uppr= list(map(lambda x:x.upper(),names))
```

```python
print(uppr)
```
