Ques. Calculate average.

```python
l= [1,2,3,4,56,6,7.8,8]

total =0

for i in l :
    total += i
avg = total/len(l)
print(avg)
```

Ques.) Count frequency of each element.

```python
lm= [1,2,3,4,5,6,6,7,8,8]
for i in set(lm) :
    
    print(i,':',lm.count(i))
```

Ques Find common elements between two lists.

```python
l1 = [1,2,3,4]
l2 = [3,4,5,6]

for i in l1:
    if i in l2 :
        print("common elements", i)
```
