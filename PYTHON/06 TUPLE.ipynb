Ques Count occurrence of an element.

```python
t = (1, 2, 3, 2, 4, 2, 5)
count=0
for i in t:
    count +=1
print(count)
```

```python
seen =[]
for i in t:
    if i not in seen:
        print(i,t.count(i))
        seen.append(i)
```

Ques .) Find max and min.

```python
t = (1, 2, 3, 2, 4, 2, 5)
max(t)
```

```python
min(t)
```

```python
largest = 0
for i in t:
    largest=i
print(largest)
```

```python
smallest = t[0]

for i in t:
    if i < smallest:
        smallest = i

print(smallest)
```
