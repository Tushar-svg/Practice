Ques. Function to check prime.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return "Its a Prime no."

is_prime(5)
```

Ques. Function to calculate factorial. 

```python
def is_fact(n):

    result = 1

    for i in range(1, n+1):
        result *=i
    
    return result

is_fact(4)
```

Ques. Function to count vowels.

```python
def total_vowel(st):

    count = 0

    for ch in st:

        if ch in ['a','e','i','o','u']:
            count+=1
    
    return count

total_vowel("Tushar")
```
