Ques. Find most frequent word in file.

```python
with open("example.txt","r") as file:
    data=file.read().lower()

    words=data.split()
    freq= {}
    for word in words:
        if word in freq:
            freq[word]+=1
        else:
            freq[word]=1
    most_freq = max(freq,key=freq.get)
    print(most_freq)
```
