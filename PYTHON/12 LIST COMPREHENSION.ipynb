Ques. Squares from 1–20.

```python
x= [x**2 for x in range(1,21)]
print(x)
```

Ques . Even numbers from list.

```python
nums = [2, 5, 8, 11, 14, 17, 20, 23, 26, 29]
even= [num for num in nums if num%2==0]
print(even)
```

Ques. Names longer than 5 chars.

```python
names = [
    "Tushar",
    "Rahul",
    "Aman",
    "Priyanka",
    "Rohit",
    "Anjali",
    "Vivek",
    "Karan",
    "Shivani",
    "Aryan"
]
```

```python
chr= [long for long in names if len(long)>5]
print(chr)
```
