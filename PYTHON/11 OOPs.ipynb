Program 1

Create: Student

Attributes: name  age marks

Methods: display() grade()

```python
class Student:
    def __init__(self,name,age,marks):
        self.name=name
        self.age=age
        self.marks=marks

    def display(self):
        return f"Details are :{self.name}, {self.age}, {self.marks}"

    def grade(self):
        if self.marks <=35:
            return "F"
        elif self.marks <= 50:
            return "E"
        elif self.marks <= 60:
            return "D"
        elif self.marks <= 75:
            return "C"
        elif self.marks <= 80:
            return "B"
        elif self.marks <= 100:
            return "A"
        else:
            return "invalid"
            
```

```python
student1 = Student("Ajay", 20, 89)
```

```python
print(student1.display())
```

```python
print(student1.grade())
```

Program 2

Create:

BankAccount

Methods:

deposit()
withdraw()
check_balance()

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance=balance

    def deposit(self,amount):
        if amount>0:
            self.__balance +=amount 

    def withdraw(self,amount):
        if amount>0:
            self.__balance -=amount

    def check_balance(self):
        return self.__balance
```

```python
amount= BankAccount(1000)
```

```python
amount.deposit(5000)
```

```python
amount.check_balance()
```

```python
amount.withdraw(3000)
```

```python
amount.check_balance()
```
