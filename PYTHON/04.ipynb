Ques.) Sum of first N numbers.

```python
n = int(input())

total = 0

for i in range(1,n+1):
    total=total+i
print(total)
```

Ques Factorial.

```python
n = int(input())
total=1
for i in range(1,n+1):
    total=total*i;
print(total)
```

Ques .) Count digits in a number.

```python
n = int(input())

count=0

while n>0:
    count+=1
    n=n//10
print(count)
```

Ques.)   Reverse a number

```python
n =int(input())
rev =0
while n>0 :
    digit= n%10
    rev = rev *10 + digit
    n= n//10
print(rev)
```

Ques.)  Armstrong number check.

```python
n = int(input())

orginal =n 
total=0
while n>0:
    digit=n%10
    n=n//10
    total+= digit **3
if orginal==total :
    print("Armstrong")
else :
    print("Not Armstrong")
```

Ques.) Fibonacci series.

```python
n = int(input())

a = 0
b = 1

for i in range(n):
    print(a, end=" ")

    c = a + b
    a = b
    b = c
```
