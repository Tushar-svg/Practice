```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds=sns.load_dataset('diamonds')
```

```python
print(diamonds.shape)
print(diamonds.head())
```

```python
diamond=diamonds[
    diamonds['cut'].isin(['Premium','Good']) &
    diamonds['color'].isin(['D','F','J'])
].sample(n=100,random_state=22)
```

```python
print(diamond.shape)
```

```python
diamond.cut=diamond.cut.cat.remove_unused_categories()
```

```python
print(diamonds.head())
```

VISUALS

```python
sns.set_style('dark')
sns.scatterplot(x='carat',y='price',data=diamond)
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(x='x',y='y',data=diamond)
plt.show()
```

SEMANTIC VARIABLES

```python
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',data=diamond)
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',palette=['purple','#55CCCC'],data=diamond)
```

3.CONTINUOUS VARIABLE 

```python
sns.scatterplot(x='carat',y='price',hue='depth',data=diamond) 
"""here DEPTH is continuous numeric variable .we are mapping this continuous
variable to the plaette """
```

4.Size

```python
sns.scatterplot(x='carat',y='price',style='cut',hue='cut',s=100,data=diamond) 
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',size='cut',data=diamond) 
#we can use sizes=[] to get the data of particular size
```

5.APLHA (lies 0 to 1 )

```python
sns.scatterplot(x='carat',y='price',style='cut',
                hue='cut',alpha=0.6,s=100,
                data=diamond) #use to avoid overlapping
```

6.MARKER

```python
sns.scatterplot(x='carat',y='price'
                ,s=100
                ,marker='*',
                edgecolor='black',
                alpha=0.4
                ,data=diamond) 
```
