```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds=sns.load_dataset('diamonds')

diamonds.head()
```

```python
diamonds=diamonds[diamonds.clarity.isin(['SI1','VS2'])]
```

```python
diamonds.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color',data=diamonds,hue='color')
```

```python
diamonds.color.value_counts(sort=False)
```

```python
sns.countplot(y='color',data=diamonds,hue='color') 
```

OPTIONS

1.Order

```python
sns.countplot(x='cut',data=diamonds,hue='cut')  #ARANGED PERFECTLY but not in upper one 
```

```python
color_order=['J','I','H','G','F','E','D']

sns.countplot(x='color',data=diamonds,hue='color',
             order=color_order)
```

Order ASCENDING or DESCENDING

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color',data=diamonds,hue='color',
             order=diamonds.color.value_counts().index)
```

HUE,HUE_ORDER

```python
sns.countplot(x='color',data=diamonds,hue='clarity',hue_order=['SI1','VS2'])
```

STYLING

color and palette

```python
sns.countplot(x='color',data=diamonds,hue='color',palette='rainbow')
```
