```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Grab parking data from UCI resource with Bash commands and read in as pandas DataFrame.

```python
df=pd.read_csv('dataset.csv',parse_dates=[3])
```

```python
df.head()
```

```python
df.size
```

```python
df=df.rename(columns={'SystemCodeNumber':'Location','LastUpdated':'Timestamp'})
```

```python
df['Day']=df.Timestamp.dt.date
df['Month']=df.Timestamp.dt.month
df['Hour']=df.Timestamp.dt.hour
```

```python
park=df[df.Location.isin(['Broad Street','NIA South'])]
```

```python
park.head()
```

VISUALS

```python
sns.set_style('dark')
```

```python
months=[pd.Timestamp(2016,10,1),
        pd.Timestamp(2016,11,1),
        pd.Timestamp(2016,12,1)]
plt.rc('date.autoformatter',day='%b %Y')  
"""rc means= Runtime Configuration
Matplotlib ki global settings change karne ke liye use hota hai.
Ye date labels ko format kar raha hai."""
```

```python
sns.lineplot(x=park.Day,y=park.Occupancy)
```

```python
sns.lineplot(x=park.Day,y=park.Occupancy)
plt.xticks(months)
```

```python
sns.lineplot(x=park.Hour,y=park.Occupancy)
```

```python
sns.lineplot(x='Day',y='Occupancy',data=park)
```

Bootstrapping

```python
sns.lineplot(x='Hour',y='Occupancy',
             n_boot=1000,
             data=park)
```

CONFIDENCE INTERVALS

```python
sns.lineplot(x='Day',y='Occupancy',
             ci=68 #default value =95 
             ,data=park)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             ci=None #its shows lowerbound and upperbound with shadding
             ,data=park)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             estimator='mean' 
             ,data=park)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             estimator='sum' 
             ,data=park)
```

hue

```python
sns.lineplot(x='Day',y='Occupancy',
             hue='Location' 
             ,data=park)
plt.xticks(months)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             hue='Location',
             palette=['gray','xkcd:brick red']
             ,data=park)
plt.xticks(months)
```

STYLE

```python
sns.lineplot(x='Day',y='Occupancy',
             style='Location' 
             ,data=park)
plt.xticks(months)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             hue='Location' ,
             style='Location'
             ,data=park)
plt.xticks(months)
```

SIZE

```python
sns.lineplot(x='Day',y='Occupancy',
             size='Location' 
             ,data=park)
plt.xticks(months)
```

```python
sns.lineplot(x='Day',y='Occupancy',
             size='Location' ,
             hue='Location'
             ,data=park)
plt.xticks(months)
```

```python
sns.lineplot(x='Hour',y='Occupancy',
             size='Month',
             ci=None
             ,data=park)
```

STYLE

```python
sns.lineplot(x='Day',y='Occupancy',
             ci=None,
             lw=4,
             color='#aa00aa',
             alpha=0.5
             ,data=park)
plt.xticks(months)
```
