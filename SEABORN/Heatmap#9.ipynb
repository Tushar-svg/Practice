```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_style('darkgrid')
```

```python
cars=sns.load_dataset('mpg')
```

```python
cars.head()
```

VISUALS

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
origin_cyl=(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
)
```

```python
sns.heatmap(origin_cyl)
```

Transpose

```python
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T)
```

Correlation

```python
cars.corr(numeric_only=True)
```

```python
sns.heatmap(cars.corr(numeric_only=True))
```

Color Palette

1. Sequential palette

```python
sns.heatmap(origin_cyl,cmap='Blues')
```

```python
2.Diverging palette
```

```python
sns.heatmap(cars.corr(numric_only=True),cmap='RdBu')
```

3.Controlling the colorbar: CENTER,VMAX,VMIN

```python
sns.heatmap(cars.corr(numeric_only=True),cmap='RdBu',center=0,vmax=1,vmin=-1)
```

Annotations

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True)
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt='.1f')
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt='.0f',
           annot_kws={
               'fontsize': 12,
               'fontweight':'bold',
               'fontfamily':'serif'
           })
```

Advance Styling

```python
sns.heatmap(cars.corr(numeric_only=True),linewidths=1,cmap='RdBu')
```

```python
sns.heatmap(cars.corr(numeric_only=True),linewidths=1,cmap='RdBu',linecolor='black')
```

Tick Labels

```python
labels=['MPS','Cylinders','Displacement','Horsepower','Weight','Acceleration','Model Year']
```

```python
sns.heatmap(cars.corr(numeric_only=True),
           cmap='RdBu',
           xticklabels=labels,
           yticklabels=labels)
```

Perfect Square

```python
sns.heatmap(cars.corr(numeric_only=True),
           cmap='RdBu',
           xticklabels=labels,
           yticklabels=labels,
           square=True)
```
