```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

VISUALS

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(x='bill_length_mm',data=penguins)
```

```python
sns.histplot(penguins.bill_length_mm)
```

```python
sns.histplot(y='bill_length_mm',data=penguins)
```

KDE (it is an another plot )

```python
sns.histplot(x='bill_length_mm'
             ,kde=True
             ,data=penguins)
```

BINS

```python
sns.histplot(x='bill_length_mm',data=penguins,bins=20,color='purple')
```

```python
sns.histplot(x='bill_length_mm',data=penguins,bins=[30,40,45,50,60]) #for specific location
```

BINWIDTH AND BINRANGE

```python
sns.histplot(x='bill_length_mm',
             binwidth=10,
             binrange=(30,60)
             ,data=penguins)
```

STATISTICS

1.stat

```python
sns.histplot(x='bill_length_mm',data=penguins
            ,stat='count')
```

a.density- AREA of histogram sums to one 

```python
sns.histplot(x='bill_length_mm',data=penguins,
            stat='density')
```

b.probability-HEIGHT of histogram bars to one

```python
sns.histplot(x='bill_length_mm',data=penguins,
            stat='probability')
```

c.cumulative -gives sense of how the probability or any stat when we moves towards x axis

```python
sns.histplot(x='bill_length_mm',data=penguins,
            stat='probability',fill=False,element='step',
            cumulative=True)
```
