```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50]
y=[0,1,2,3,4]
```

```python
plt.plot(y,x)
plt.show()
```

```python
sns.set_theme()  
sns.set_style("dark")  #five built-in themes/styles: darkgrid, whitegrid, dark, white, and ticks
plt.plot(y,x)
plt.show()
```

```python
sns.relplot(x='model_year',y='mpg',col='origin',hue='cylinders',data=cars) #btw THIS IS NOT WE ARE ITS JUST AN EXAMPLE
plt.show()
```
