```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

VISUALS

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(tips)
```

PLOT KINDS

diag_kind

```python
sns.pairplot(tips,diag_kind='kde')
```

kind

```python
sns.pairplot(tips,kind='reg')
```

Returns PairGrid

```python
g=sns.pairplot(tips)
```

```python
type(g)
```

```python
g=sns.pairplot(tips)
g.map_upper(sns.kdeplot,n_levels=6,color='xkcd:wine red')
```

CATEGORICAL & SPECIFIC VARIABLES

```python
tips['weekend']=tips.day.isin(['Sat','Sun'])
```

```python
tips.head()
```

hue

```python
sns.pairplot(tips,hue='weekend')
```

vars,x_vars,y_vars

```python
sns.pairplot(tips,hue='weekend',vars=['total_bill','tip'])
```

```python
sns.pairplot(tips,hue='weekend',
            y_vars=['tip'],
            x_vars=['total_bill','size'],
            kind='reg',
            height=4)
```

ADVANCES STYLING

height,aspect

```python
from ipywidgets import interactive

def size_widget(height=2.5,aspect=1):
    sns.pairplot(tips,height=height,aspect=aspect)
```

```python
interactive(size_widget,height=(1,3.5,0.5),aspect=(0.5,2,0.25))
```

palette

```python
sns.pairplot(tips,hue='day',palette='plasma')
```

diag_kws,plot_kws

```python
sns.pairplot(tips,diag_kws={'color':'gray','alpha':0.6})
```

```python
sns.pairplot(tips,
            kind='reg',
            plot_kws={'ci':None,'color':'gray',
                     'scatter_kws':{'color':'xkcd:wine red'}})
```
