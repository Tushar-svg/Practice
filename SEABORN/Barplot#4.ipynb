```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np 
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

Visuals

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species'
            ,data=penguins)
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(x='body_mass_g',y='species',hue='species',data=penguins)
```

CoNFIDENCE INTERVALS (box uppr ticks)

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species'
            ,ci=95
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species'
            ,ci=68
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species',
            ci=None
            ,data=penguins)
```

ESTIMATOR

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species',
            estimator=np.std # estimator='std' also works here
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g'
            ,hue='species',
            estimator=np.max
            ,data=penguins)
```

CATEGORICAL VARIABLES

```python
sns.barplot(x='species',y='body_mass_g',
            hue='sex',
            palette=['green','xkcd:brick red']
            ,data=penguins)
```

```python
sns.barplot(x='body_mass_g',y='species',
            hue='sex',
            palette=['green','xkcd:brick red']
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g',
            order=['Gentoo','Chinstrap','Adelie'],
            hue='species'
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g',
            order=['Gentoo','Chinstrap','Adelie'],
            hue='sex',
            hue_order=['Female','Male']
            ,data=penguins)
```

STYLING COLORS

```python
sns.barplot(x='species',y='body_mass_g',
            color='c'
            ,data=penguins)
```

errcolor & errwidth

```python
sns.barplot(x='species',y='body_mass_g',
            color='c',
            errcolor='darkslateblue'
            ,errwidth=8
            ,data=penguins)
```

```python
sns.barplot(x='species',y='body_mass_g',
            palette='hot'
            ,data=penguins)
```

edgecolor

```python
sns.barplot(x='species',y='body_mass_g',ci=None,
            color='aliceblue',
            edgecolor='mediumblue',
            lw=3
            ,data=penguins)
```
