```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
 cars=sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

VISUALS

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.mpg)
```

```python
sns.boxplot(x=cars.origin,y=cars.mpg)
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,
           hue='origin')
```

```python
#above graph 
```

1.BY cylinders

```python
sns.boxplot(x='origin',y='mpg',data=cars,
           hue='cylinders')
```

2.BY MODEL YEAR

```python
cars.model_year.describe()
```

```python
cars['newer_model']=cars.model_year>76
```

```python
sns.boxplot(x='origin',y='mpg',data=cars,
           hue='newer_model')
```

STYLING

1.Order

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin')
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='newer_model',
           order=['japan','europe','usa'],
           hue_order=[True,False])
```

2.color

```python
sns.boxplot(x='mpg',y='origin',data=cars,hue='newer_model',
           color='green')
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,hue='newer_model',
           palette='dark:green')
```

3.width

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           width=0.3)
```

4.linewidth

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           linewidth=2.7)
```

5.whis

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           whis=1)
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           whis=2)
```

6.fliersize

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           fliersize=9)
```

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           fliersize=3)
```

7.showcaps

```python
sns.boxplot(x='mpg',y='origin',data=cars,
           hue='origin',
           showcaps=False)
```
