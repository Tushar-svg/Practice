```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years,runs)
plt.xlabel('Years')
plt.ylabel('Runs Scored ')
plt.title('Yearly Runs scored by Sachin Tendulkar')
plt.tight_layout()
plt.show()
```

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x=np.arange(len(years))
width=0.25
```

```python
plt.bar(x-width,sachin,width=width,color='red',label='Sachin Tendulkar',alpha=0.6)
plt.bar(x,sehwag,width=width,color='blue',label='Sehwag',alpha=0.6)
plt.bar(x+width,kohli,width=width,color='green',label='Virat Kohli',alpha=0.6)
plt.xlabel('Years')
plt.ylabel('Runs Scored ')
plt.title('Yearly Runs scored by Sachin Tendulkar')
plt.xticks(x,years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100,
300+600+800+1100+900]

plt.bar(players,runs_5yrs,color='skyblue')
for i in range(len(players)):
    plt.text(i, runs_5yrs[i] + 50, str(runs_5yrs[i]), ha='center')
plt.xlabel("Total Runs in First 5 Years")
plt.ylabel("Players")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```

```python
run = [1500, 1200, 1800,1600]
plt.bar(players,run,color='magenta',alpha=0.7)
for i in range(len(players)):
    plt.text(i, run[i] + 50, str(run[i]), ha='center')
plt.xlabel("Total Runs in First 5 Years")
plt.ylabel("Players")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```
