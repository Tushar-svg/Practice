```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
```

```python
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

```python
plt.style.use('ggplot')
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
```

```python
plt.pie(runs,labels=labels,colors=colors)
```

```python
plt.pie(runs,
        labels=labels,
        colors=colors,
       wedgeprops={'edgecolor':'black','linestyle':'--','linewidth':2})
```

```python
explode = [0.05, 0, 0, 0]
plt.pie(runs,
        labels=labels,
        colors=colors,
       wedgeprops={'edgecolor':'black','linestyle':'--','linewidth':2},
       explode=explode)
```

```python
explode = [0.1, 0, 0, 0]
plt.pie(runs,
        labels=labels,
        colors=colors,
       
       explode=explode,
       shadow=True,
       )
```

```python
plt.pie(runs,
        labels=labels,
        colors=colors,
       startangle=140,
       explode=explode,
       shadow=True,
       )
```

```python
plt.pie(runs,
        labels=labels,
        colors=colors,
       startangle=140,
       explode=explode,
        wedgeprops={'edgecolor':'black','linestyle':'--','linewidth':0.8},
       shadow=True,
        autopct='%1.1f%%'      #Avoid using it in too much categories
       )
```
