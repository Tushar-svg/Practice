```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years,runs)
plt.title('Sachin Yearly scores')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.tight_layout()
plt.grid(True)
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years,runs,'black',linestyle='-.',label='Sachin Tendulkar')
plt.plot(years,kohli,'bo--',label='Virat kohli')
plt.plot(years,sehwag,'g^:',linewidth=3,label='Sehwag')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison')
plt.tight_layout()
plt.legend()
plt.grid(True)
```

```python
with plt.xkcd():
    plt.plot(years,runs,'black',label='Sachin Tendulkar')
    plt.plot(years,kohli,'bo--',label='Virat kohli')
    plt.plot(years,sehwag,'g^:',label='Sehwag')
    plt.xlabel('Year')
    plt.ylabel('Runs Scored')
    plt.title('Performance Comparison')
    plt.tight_layout()
    plt.legend()
    plt.grid(True)
```

```python
ages=df['Age']
devs_salary=df['All_Devs']
py_salary=df['Python']
js_salary=df['JavaScript']
```

```python
plt.style.use('fivethirtyeight')
plt.plot(ages,devs_salary,color='indigo',linestyle='-.',label='All Devs')
plt.plot(ages,py_salary,color='red',label='Python')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.style.use('classic')
plt.plot(ages,devs_salary,color='red',linestyle='-',linewidth=1.5,label='All Devs')
plt.plot(ages,py_salary,color='black',label='Python')
plt.fill_between(ages,py_salary,alpha=0.25)
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.style.use('Solarize_Light2')
plt.plot(ages,devs_salary,color='red',linestyle='-',linewidth=1.5,label='All Devs')
plt.plot(ages,py_salary,color='black',label='Python')
overall_median=57287
plt.fill_between(ages,py_salary,overall_median,alpha=0.25)
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages,devs_salary,color='red',linestyle='-',linewidth=1.5,label='All Devs')
plt.plot(ages,py_salary,color='black',label='Python')
overall_median=57287
plt.fill_between(ages,py_salary,overall_median,alpha=0.25,
                where=(py_salary > overall_median),
                interpolate=True)
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages,devs_salary,color='red',linestyle='-',linewidth=1.5,label='All Devs')
plt.plot(ages,py_salary,color='indigo',label='Python')
overall_median=57287
plt.fill_between(ages,py_salary,overall_median,alpha=0.25,
                where=(py_salary > overall_median),
                interpolate=True,color='green')

plt.fill_between(ages,py_salary,overall_median,alpha=0.25,
                where=(py_salary <= overall_median),
                interpolate=True,color='red')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages,devs_salary,color='red',linestyle='-',linewidth=1.5,label='All Devs')
plt.plot(ages,py_salary,color='indigo',linewidth=1.5,label='Python')
overall_median=57287
plt.fill_between(ages,py_salary,devs_salary,alpha=0.25,
                where=(py_salary > devs_salary),
                interpolate=True,label='Above Avg.')

plt.fill_between(ages,py_salary,devs_salary,alpha=0.25,
                where=(py_salary <= devs_salary),
                interpolate=True,color='red',label='Below Avg.')

plt.legend(loc='best') #'upper left', 'upper right', 'lower left', 'lower right','center left', 'center right', 'lower center', 'upper center', 'center'
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
from datetime import datetime,timedelta
import matplotlib.dates as mpl_date
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot(dates,y,'o',linestyle='solid')
plt.tight_layout()
plt.show()
```

```python
plt.plot(dates,y,'o',linestyle='solid')
plt.gcf().autofmt_xdate()  #GCF= GET CURRENT FIGURE

plt.tight_layout()
plt.show()
```

```python
plt.plot(dates,y,'o',linestyle='solid')

plt.gcf().autofmt_xdate()  #GCF= GET CURRENT FIGURE

date_format= mpl_date.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_format)  #gca= GET CURRENT AXIS

plt.tight_layout()
plt.show()
```

```python
price_date=data['Date']
price_close=data['Close']
```

```python
plt.plot(price_date,price_close,'o',linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.tight_layout()
plt.show()
```

```python
data['Date']=pd.to_datetime(data['Date'])
data.sort_values('Date',inplace=True)

price_date=data['Date']
price_close=data['Close']
plt.plot(price_date,price_close,'o',linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.tight_layout()
plt.show()
```
